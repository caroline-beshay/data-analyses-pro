```python
from sklearn.preprocessing import LabelEncoder
encoding=LabelEncoder()
car['transmission']=encoding.fit_transform(car['transmission'])
car['fuelType']=encoding.fit_transform(car['fuelType'])
car['categoral car']=encoding.fit_transform(car['categoral car'])
car['year']=encoding.fit_transform(car['year'])
car.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
car['fuelType']=scaler.fit_transform(car[['fuelType']])
car['transmission']=scaler.fit_transform(car[['transmission']])
car['categoral car']=scaler.fit_transform(car[['categoral car']])
car['year']=encoding.fit_transform(car['year'])
```

```python
car.head()
```

```python
x=car['total price']
y=car.iloc[:,[2,3,4,5,6,8]]
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
selector=SelectKBest(score_func=f_regression,k=5)
xneaw=selector.fit_transform(y,x)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(xneaw,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
r2_score=r2_score(ypred,ytest)
mae=mean_absolute_error(ypred,ytest)
```
